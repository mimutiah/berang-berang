### **<span style='color:#E1B12D'><i>Binned Age </i></span>**



**Kolom Baru**: `age_bin`

**Deskripsi**: Mengelompokkan usia ke dalam kategori sebagai berikut:
- **<30**: Usia di bawah 30 tahun
- **30-40**: Usia antara 30 hingga 39 tahun
- **40-50**: Usia antara 40 hingga 49 tahun
- **50-60**: Usia antara 50 hingga 59 tahun
- **>=60**: Usia 60 tahun ke atas

**Alasan**: Pengelompokan usia ini dirancang untuk memudahkan analisis pola perilaku yang berbeda di antara kelompok usia. Dengan kategori ini, kita dapat lebih mudah memahami perbedaan dalam preferensi atau respons terhadap kampanye deposito bank berdasarkan usia. Ini juga membantu dalam menyederhanakan data untuk analisis yang lebih mendalam, serta memfasilitasi identifikasi segmen pasar yang lebih relevan.


```python
import pandas as pd

# Definisikan bin dan label sesuai dengan kategori usia
bins = [17, 29, 39, 49, 59, 100]  # Rentang usia
labels = ['Muda', 'Dewasa', 'Dewasa Tua', 'Tua', 'Lansia']  # Label untuk bin

# Tambahkan kolom 'age_bin' berdasarkan bin
df2['age_bin'] = pd.cut(df2['age'], bins=bins, labels=labels, right=True)

# Tampilkan beberapa baris pertama untuk memeriksa hasil
print(df2[['age', 'age_bin']].head())
```

### **<span style='color:#E1B12D'><i>Duration previous interaction </i></span>**

**Kolom Baru: duration_prev_interaction**

- **Deskripsi**: Fitur ini merupakan rasio antara durasi kontak terakhir (`duration`) dan jumlah kontak sebelumnya (`previous`). Dalam perhitungannya, kita menambahkan angka 1 pada kolom `previous` untuk menghindari pembagian dengan nol, yang dapat menyebabkan error dalam perhitungan.

- **Alasan**: Fitur ini dirancang untuk mengukur seberapa efektif kontak sebelumnya dalam memperpanjang durasi interaksi saat ini. Sebagai contoh, jika durasi kontak terakhir cukup lama namun jumlah kontak sebelumnya rendah, rasio ini akan tinggi, menunjukkan bahwa kontak sebelumnya mungkin sangat efektif dalam menjaga interaksi yang lebih panjang. Sebaliknya, rasio yang rendah bisa menunjukkan bahwa kontak sebelumnya kurang berhasil dalam memperpanjang durasi interaksi.

```python
# Menambahkan fitur interaksi
df2['duration_prev_interaction'] = df2['duration'] / (df2['previous'] + 1)  # menambahkan 1 untuk menghindari pembagian dengan nol
```

### **<span style='color:#E1B12D'><i>Correlation Between Economic Factors </i></span>**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan df2 adalah DataFrame Anda
# Pilih kolom-kolom yang relevan
columns_of_interest = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

# Pilih subset data
df_subset = df2[columns_of_interest]

# Hitung matriks korelasi
correlation_matrix = df_subset.corr()

# Buat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Heatmap Korelasi Faktor Ekonomi')
plt.show()
```

Dari heatmap, terlihat adanya korelasi yang signifikan antara beberapa faktor ekonomi. Korelasi ini wajar karena faktor-faktor ekonomi seringkali saling memengaruhi. Misalnya, tingkat variasi pekerjaan (`emp.var.rate`) dapat mempengaruhi indeks harga konsumen (`cons.price.idx`) karena perubahan besar dalam pekerjaan dapat memengaruhi daya beli dan harga barang. Selain itu, tingkat euribor 3 bulan (`euribor3m`) mempengaruhi biaya pinjaman dan investasi, yang berdampak pada jumlah karyawan (`nr.employed`), dan perubahan dalam jumlah karyawan dapat mempengaruhi variasi pekerjaan serta keterkaitan lainnya yang akan dibahas lebih lanjut.

Namun, dalam konteks pemodelan, korelasi tinggi antar faktor ini dapat menimbulkan masalah multikolinearitas, di mana beberapa variabel independen sangat berkorelasi satu sama lain, sehingga mempengaruhi kestabilan model. Oleh karena itu, dalam pemodelan, faktor-faktor ekonomi yang memiliki korelasi tinggi mungkin perlu dihapus untuk menghindari masalah ini. Namun, untuk analisis bisnis sementara ini, semua faktor ekonomi dibiarkan untuk memahami dampak dan hubungan antar variabel dalam konteks yang lebih luas.

**Kolom Baru: `emp.var.rate_bin`, `cons.price.idx_bin`, `cons.conf.idx_bin`, `euribor3m_bin`, `nr.employed_bin`**

**Deskripsi**: Fitur-fitur ini mengelompokkan nilai dari kolom-kolom `emp.var.rate`, `cons.price.idx`, `cons.conf.idx`, `euribor3m`, dan `nr.employed` ke dalam kategori rentang tertentu.

**Alasan**: Binning ini menyederhanakan analisis dengan mengelompokkan nilai kontinu menjadi kategori yang lebih mudah dikelola, memudahkan pemahaman dampak dari rentang nilai tertentu terhadap hasil yang diinginkan.

```python
import pandas as pd
import numpy as np
from IPython.display import display

# Misalkan df2 adalah dataset kamu

# Tentukan rentang untuk setiap kolom
bins_emp_var_rate = [-np.inf, -2, 0, 1.1, np.inf]
labels_emp_var_rate = ['< -2', '-2 - 0', '0 - 1.1', '> 1.1']

bins_cons_price_idx = [92, 93, 94.5, 95]
labels_cons_price_idx = ['92 - 93', '93 - 94.5', '94.5 - 95']

bins_cons_conf_idx = [-60, -50, -40, -30]
labels_cons_conf_idx = ['-60 - (-50)', '-50 - (-40)', '-40 - (-30)']

bins_euribor3m = [0, 1, 4.5, 5]
labels_euribor3m = ['0 - 1', '1 - 4.5', '4.5 - 5']

bins_nr_employed = [5000, 5100, 5200, np.inf]
labels_nr_employed = ['5000 - 5100', '5100 - 5200', '> 5200']

# Melakukan binning pada setiap kolom dengan batas atas inklusif (right=True)
df2['emp.var.rate_bin'] = pd.cut(df2['emp.var.rate'], bins=bins_emp_var_rate, labels=labels_emp_var_rate, right=True, include_lowest=True)
df2['cons.price.idx_bin'] = pd.cut(df2['cons.price.idx'], bins=bins_cons_price_idx, labels=labels_cons_price_idx, right=True, include_lowest=True)
df2['cons.conf.idx_bin'] = pd.cut(df2['cons.conf.idx'], bins=bins_cons_conf_idx, labels=labels_cons_conf_idx, right=True, include_lowest=True)
df2['euribor3m_bin'] = pd.cut(df2['euribor3m'], bins=bins_euribor3m, labels=labels_euribor3m, right=True, include_lowest=True)
df2['nr.employed_bin'] = pd.cut(df2['nr.employed'], bins=bins_nr_employed, labels=labels_nr_employed, right=True, include_lowest=True)

# Menampilkan hasilnya dengan display
display(df2[['emp.var.rate', 'emp.var.rate_bin']].head())
display(df2[['cons.price.idx', 'cons.price.idx_bin']].head())
display(df2[['cons.conf.idx', 'cons.conf.idx_bin']].head())
display(df2[['euribor3m', 'euribor3m_bin']].head())
display(df2[['nr.employed', 'nr.employed_bin']].head())
```
